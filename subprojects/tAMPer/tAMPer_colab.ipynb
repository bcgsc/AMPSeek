# **tAMPer**: structure-aware deep learning model for toxicity prediction of anti-microbial peptides (AMPs)

Easy to use toxicity prediction tool of peptides. For details, refer to our manuscript.

**Usage**

`input_dir` directory with only the fasta file (`.fasta` format) stored in Google Drive.

`pdbs_dir` directory with predicted structures in Google Drive. Structures need to be PDB formatted and have an `.pdb` for one structure per sequence or `.result.gz` which has 5 predicted structures from colabFold.

`result_dir` results will be written to the result directory in Google Drive.
